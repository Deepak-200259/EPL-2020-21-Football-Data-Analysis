```python
#Importing Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Dataset

epl_df=pd.read_csv('D:/Project/Football EPL 2020-21 Data Analysis in Python/Dataset/EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info() #No null columns Found
```

```python
epl_df.describe()
```

```python
epl_df.isnull().sum()
```

```python
#create 2 new columns - mins per match and goals per match

epl_df['MinsPerMatch']=(epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']=(epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
#Total Goals scored in season
Total_Goals = epl_df['Goals'].sum()
Total_Goals
```

```python
#Total goals scroed in penalties in season
Penalty_goals = epl_df['Penalty_Goals'].sum()
Penalty_goals
```

```python
#Total Penalty Attempted in season
Attempted_penalties = epl_df['Penalty_Attempted'].sum()
Attempted_penalties
```

```python
#Total number of Missed Penalties or Blocked penalty in season
Penalty_saves= Attempted_penalties - Penalty_goals
Penalty_saves
```

```python
#pie chart for penalties missed and penalties scored
plt.figure(figsize=(13,6))
data = [Penalty_saves, Penalty_goals]
labels = ['Penalties Missed','Penalties Scored']
color = sns.color_palette('Set1')
plt.pie(data,labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
#Top 5 Clubs with maximum player in their squad
top_clubs = epl_df['Club'].value_counts().sort_values(ascending=False)
top_clubs.head(5).plot(kind='bar',color=sns.color_palette('magma'))
# OR  epl_df['Club'].value_counts().nlargest(5).plot(kind='bar',color=sns.color_palette('magma'))
```

```python
# Players based on age group
Under20 = epl_df[epl_df['Age']<=20]
Under25 = epl_df[(epl_df['Age']<=25) & (epl_df['Age']>20)]
Under30 = epl_df[(epl_df['Age']<=30) & (epl_df['Age']>25)]
Above30 = epl_df[epl_df['Age']>30]
```

```python
#Pie Chart of different age groups
x=np.array([Under20['Name'].count(),Under25['Name'].count(),Under30['Name'].count(),Above30['Name'].count()])
mylabels = ["Less than 20","20 to 25", "25 to 30", "More than 30"]
plt.title('Total Players with Age Groups',fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
#Total Player under 20 in each club
players_under20 = epl_df[epl_df['Age']<20]
players_under20['Club'].value_counts().plot(kind='bar',color=sns.color_palette('cubehelix'))
```

```python
#Average age of players in each club
plt.figure(figsize=(14,9))
sns.boxplot(x='Club',y='Age',data=epl_df)
plt.xticks(rotation=90)
```

```python
num_Player = epl_df.groupby('Club').size()
data= (epl_df.groupby('Club')['Age'].sum())/num_Player
data.sort_values(ascending=False)

#Conclusion - 
#1. The Average age of Crystal Palace club is 28.33 which is highest
#2. The Average age of Manchestar United club is 23.86 which is Least
#3. Crystal Palace has more number of Aged players and manchestar United has more number of Young Players
```

```python
#Total Assists from Each Club
Assists_by_club= pd.DataFrame(epl_df.groupby('Club',as_index=False)['Assists'].sum())
ax=sns.barplot(x='Club',y='Assists',data=Assists_by_club.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,10)
plt.title('Plot of Clubs vs Total Assists',fontsize=20)
```

```python
Assists_by_club= pd.DataFrame(epl_df.groupby('Club',as_index=False)['Assists'].sum())
Assists_by_club.sort_values(by='Assists',ascending=False)
#Conclusion - 
#1. Manchester City has the Highest number of Assists in the season i.e. 55
#2. Sheffield United has the Lowest Number of Assists in the season i.e. 13
```
